# Scacchiera

```python
def creo_scacchiera(n):                       #definisco una funzione che mi crea la scacchiera
    scacchiera = np.zeros((n*2+1,n*2+1))      #la mia scacchiera vuota
    x = n;y= n                                #centro
    scacchiera[x][y] = 1                      #imposto il centro = 1

    count = 1                                 #numero da cui parte la scacchiera
    for circle in range (n):                  #faccio partire il ciclo
        passi = 2 * (circle +1)
        x += 1
        y+= 1
        for go_up in range (passi):           #spostamento in alto
            x-=1                              #RICORDA CHE X e Y SONO LE RIGHE E COLONNE (non immaginarlo come un piano cartesiano)
            count += 1
            scacchiera [x][y]  = count
    
        for go_left in range (passi):         #spostamento a sinistra
            y-=1
            count += 1
            scacchiera [x][y] = count

        for go_down in range(passi):          #spostamento in basso
            x += 1
            count += 1
            scacchiera [x][y] = count
    
        for go_right in range (passi):        #spostamento a destra
            y += 1
            count += 1
            scacchiera [x][y] = count


    return scacchiera
    
```

Walk

```python
def simulate_walk(scacchiera):
    valori=[]; coordinate_x = []; coordinate_y = []
    x = 0; y= 0
    spostamenti = ((x+2,y-1), (x+2,y+1),(x-2,y+1),(x-2,x-1),(x+1,y+2),(x+1,y-2),(x-1,y+2),(x-1,y-2)) #lista degli spostamenti possibili del cavallo
    # creo una matrice di zeri per salvare i movimenti
   
    len_scacchiera = np.size(scacchiera,0)   #ho la lunghezza della scacchiera
    max_pos = len_scacchiera//2              #ricavo il centro (doppio slash per l'approsimazione)
    visitati = np.zeros ((len_scacchiera, len_scacchiera),bool)

    #ottengo la posizione centrale
    visitati[y+max_pos, x+max_pos] = True
    valori.append(scacchiera[y+max_pos,x+max_pos]); coordinate_x.append(x); coordinate_y.append(y)    #salvare i valori nelle liste

    #MUOVIAML inizia il ciclo
    fine_gioco = False
    while not fine_gioco:
        board_min = pow(len_scacchiera,2)
        fine_gioco = True
        spostamento = ((x+2,y-1), (x+2,y+1),(x-2,y+1),(x-2,x-1),(x+1,y+2),(x+1,y-2),(x-1,y+2),(x-1,y-2))
    #check delle due scacchiere
        for test_x, test_y in spostamento:
            #abs restituisce il valore assoluto di un numero
    
            if abs(test_x)<= max_pos and abs(test_y) <= max_pos and not visitati[test_y+max_pos,test_x+max_pos]:
                test_scacchiera = scacchiera[test_y+max_pos, test_x+max_pos]
                if test_scacchiera < board_min:
                    fine_gioco = False
                    board_min = test_scacchiera; x = test_x; y = test_y

        if not fine_gioco:
            visitati[y+max_pos,x+max_pos] = True
            valori.append(board_min); coordinate_x.append(x); coordinate_y.append(y)

    return(valori, coordinate_x, coordinate_y)
```

```python
scacchiera = creo_scacchiera(3)
display(scacchiera)

valori, coordinate_x, coordinate_y = simulate_walk(scacchiera)
print ("Sequence = ", valori)
print ("Coordinate x = ", coordinate_x)
print ("coordinate y =", coordinate_y)
```
